```python
import pandas as pd
import random
import csv
import numpy as np
from pandas.api.types import is_string_dtype
```

```python
recipe_csv = pd.read_csv("recipe.csv")
recipe_has_eq_csv = pd.read_csv("recipe_has_equipment.csv")
recipe_has_ing_csv = pd.read_csv("recipe_has_ingredient.csv")
recipe_has_label_csv = pd.read_csv("recipe_has_label.csv")
recipe_has_meal_csv = pd.read_csv("recipe_has_meal.csv")
recipe_has_tip_csv = pd.read_csv("recipe_has_tip.csv")
step_csv = pd.read_csv("step.csv")

chef_data = pd.read_csv("../data/chef.csv")
```

```python
# Append new cuisines
cur_cui = pd.read_csv("../data/to_import/cuisine.csv").to_dict(orient='list')

for row in recipe_csv.itertuples(index=True, name='Pandas'):
    cuisine_name = getattr(row, "cuisine_name")
    if not(cuisine_name in cur_cui["name"]):
        cur_cui["name"].append(cuisine_name)
        cur_cui["image_id"].append(cur_cui["image_id"][-1]+1)

print(cur_cui)
pd.DataFrame(cur_cui).to_csv("generated/cuisine.csv", index=False, quoting=csv.QUOTE_NONNUMERIC, quotechar='"')
```

```python
# Modify recipes

# Change column names
new_recipe = recipe_csv.rename(columns={'cook_time': 'cook_time_m', 'prep_time': 'prep_time_m'})
# Change type names
new_recipe['type'] = np.where(new_recipe['type'] == 'meal', 'Cooking', 'Pastry')

old_recipe = pd.read_csv("../data/recipe.csv").rename(columns={'cook_time': 'cook_time_m', 'prep_time': 'prep_time_m'})
old_recipe['type'] = np.where(old_recipe['type'] == 'meal', 'Cooking', 'Pastry')

frames = [new_recipe, old_recipe]

# all_recipe is now the DataFrame that contains all of our recipes
all_recipe = pd.concat(frames)

all_recipe.to_csv("generated/recipe.csv", index=False, quoting=csv.QUOTE_NONNUMERIC, quotechar='"')
```

```python
# chef_has_cuisine generator

chef_has_cuisine = {
    "chef_id":[],
    "cuisine_name":[]
}

chef_ids = []
for row in chef_data.itertuples(index=True, name='Pandas'):
    chef_ids.append(getattr(row, 'id'))

for id in chef_ids:
    s = random.randint(1, 3)
    selection = random.sample(cur_cui["name"], s)
    
    for i in range(0, s):
        chef_has_cuisine["chef_id"].append(id)       
        chef_has_cuisine["cuisine_name"].append(selection[i])

pd.DataFrame(chef_has_cuisine).to_csv("generated/chef_has_cuisine.csv", index=False, quoting=csv.QUOTE_NONNUMERIC, quotechar='"')
```

```python
recipe_cuisine = {}
for row in all_recipe.itertuples(index=True, name='Pandas'):
    recipe_name = getattr(row, 'name')
    cui_name = getattr(row, 'cuisine_name')
    if recipe_name in recipe_cuisine.keys(): 
        recipe_cuisine[recipe_name].append(cui_name)
    else:
        recipe_cuisine[recipe_name] = [cui_name]
#print(recipe_cuisine)
```

```python
# recipe_has_chef generator

recipe_has_chef = {
    "recipe_name": [],
    "recipe_cuisine_name": [],
    "chef_id": []
}

# key: cuisine, value: list of chef ids that has that cuisine
rcuisine = {}
for i in range(0, len(chef_has_cuisine["chef_id"])):
    c = chef_has_cuisine["cuisine_name"][i]
    id = chef_has_cuisine["chef_id"][i]
    if not(c in rcuisine.keys()):
        rcuisine[c] = [id]
    else:
        rcuisine[c].append(id)
#print(rcuisine)

# key: cuisine, value: list of recipes that are of this cuisine
rrecipe = {}
for lr in recipe_cuisine.keys():
    for lc in recipe_cuisine[lr]:
        if not(lc in rrecipe.keys()):
            rrecipe[lc] = [lr]
        else:
            rrecipe[lc].append(lr)
#print(rrecipe)

for cuisine in rcuisine.keys():
    for id in rcuisine[cuisine]:
        for rec in rrecipe[cuisine]:
            recipe_has_chef["recipe_name"].append(rec)
            recipe_has_chef["recipe_cuisine_name"].append(cuisine)
            recipe_has_chef["chef_id"].append(id)

pd.DataFrame(recipe_has_chef).to_csv("generated/recipe_has_chef.csv", index=False, quoting=csv.QUOTE_NONNUMERIC, quotechar='"')
```

```python
# step editor
step_data = pd.read_csv("../data/step.csv") # old step data
# remove ukrainian recipe steps
step_data = step_data[step_data.recipe_cuisine_name != 'Ukrainian']

step_frames = [step_csv.rename(columns={'name': 'recipe_name', 'cuisine_name': 'recipe_cuisine_name'}), step_data]

all_step = pd.concat(step_frames)
all_step.to_csv("generated/step.csv", index=False, quoting=csv.QUOTE_NONNUMERIC, quotechar='"')
```

```python
# change _has_equipment cuisines and create equipment
he = pd.read_csv("../data/recipe_has_equipment.csv")
he = he[he.recipe_cuisine_name != 'Ukrainian']

equipment = {
    "name":[],
    "manual":[],
    "image_id":[]
}

rhe_frames = [recipe_has_eq_csv, he]
all_rhe = pd.concat(rhe_frames)
all_rhe = all_rhe.rename(columns={'equipment':'equipment_name'})
all_rhe.to_csv("generated/recipe_has_equipment.csv", index=False, quoting=csv.QUOTE_NONNUMERIC, quotechar='"')

for row in all_rhe.itertuples(index=True, name='Pandas'):
    recipe_name = getattr(row, 'recipe_name')
    
    eq = getattr(row, 'equipment_name')
    if not(eq in equipment["name"]):
        equipment["name"].append(eq)
        equipment["manual"].append("used for cooking")
        equipment["image_id"].append(0)

pd.DataFrame(equipment).to_csv("generated/equipment.csv", index=False, quoting=csv.QUOTE_NONNUMERIC, quotechar='"')
```

```python
food_team = pd.read_csv("../data/aftermath/food_team.csv")
short_names = food_team["short_name"].to_numpy()
```

```python
# ingredient things
basic_ingredients = []
for row in all_recipe.itertuples(index=True, name='Pandas'):
    ing = getattr(row, 'ingredient_name')
    if not(ing in basic_ingredients):
        basic_ingredients.append(ing)

food_info = {
    "fat":[],
    "protein":[],
    "carbohydrate":[],
    "recipe_name":[],
    "recipe_cuisine_name":[]
}

ingredient = {
    "name":[],
    "cal_per_100":[],
    "food_team_short_name":[],
    "image_id":[]
}

hi = pd.read_csv("../data/recipe_has_ingredient.csv")
hi_frames = [recipe_has_ing_csv, hi]
hi = pd.concat(hi_frames) # all_hi 

for row in hi.itertuples(index=True, name='Pandas'):
    recipe_name = getattr(row, 'recipe_name')
    
    ing = getattr(row, 'ingredient_name')
    if not(ing in ingredient["name"]):
        ingredient["name"].append(ing)
        w = getattr(row, "ingredient_weight")
        cal_per = getattr(row, "ingredient_kcal")
        if w > 0:
            cal_per = int(cal_per/w) * 100
        ingredient["cal_per_100"].append(cal_per)
        # RANDOM FOOD TEAM SHORT NAME
        ingredient["food_team_short_name"].append(random.choice(short_names))
        ingredient["image_id"].append(0)

    if not(recipe_name in food_info["recipe_name"]):
        for rc in recipe_cuisine[recipe_name]:
            food_info["recipe_name"].append(recipe_name)
            food_info["recipe_cuisine_name"].append(rc)
            food_info["carbohydrate"].append(int(getattr(row, "ingredient_carbs")))
            food_info["fat"].append(random.randint(0, 50))
            food_info["protein"].append(random.randint(0, 50))
    else:
        food_info["carbohydrate"][food_info["recipe_name"].index(recipe_name)] += int(getattr(row, "ingredient_carbs"))
        
for ing in basic_ingredients:
    if not(ing in ingredient["name"]):
        print("found a fker: {0}".format(ing))
        ingredient["name"].append(ing)
        ingredient["cal_per_100"].append(20)
        ingredient["food_team_short_name"].append(random.choice(short_names))
        ingredient["image_id"].append(0)


new_hi = hi.rename(columns={'ingredient_weight': 'quantity'})
new_hi = new_hi.drop(columns=['ingredient_full_name', 'ingredient_kcal', 'ingredient_carbs'])
new_hi.quantity = new_hi.quantity.astype(int)
new_hi = new_hi.drop_duplicates(subset=['recipe_name', 'recipe_cuisine_name', 'ingredient_name'])
new_hi.to_csv("generated/recipe_has_ingredient.csv", index=False, quoting=csv.QUOTE_NONNUMERIC, quotechar='"')

pd.DataFrame(food_info).to_csv("generated/food_info.csv", index=False, quoting=csv.QUOTE_NONNUMERIC, quotechar='"')

pd.DataFrame(ingredient).to_csv("generated/ingredient.csv", index=False, quoting=csv.QUOTE_NONNUMERIC, quotechar='"')
```

```python
# Generate label and change _has_label
hl = pd.read_csv("../data/recipe_has_label.csv")
all_hl = pd.concat([recipe_has_label_csv, hl])
all_hl.to_csv("generated/recipe_has_label.csv", index=False, quoting=csv.QUOTE_NONNUMERIC, quotechar='"')
```

```python
# Generate meal and change _has_meal
hm = pd.read_csv("../data/recipe_has_meal.csv")
all_hm = pd.concat([recipe_has_meal_csv, hm])
all_hm.to_csv("generated/recipe_has_meal.csv", index=False, quoting=csv.QUOTE_NONNUMERIC, quotechar='"')
```

```python
# Generate tip and change _has_tip
ht = pd.read_csv("../data/recipe_has_tip.csv")
all_ht = pd.concat([recipe_has_tip_csv, ht])

tip = {
    "id":[],
    "tip":[]
}
ids = []
for row in all_ht.itertuples(index=True, name='Pandas'):
    recipe_name = getattr(row, 'recipe_name')

    tip_tip = getattr(row, "tip")
    if not tip["id"]:
        tip["id"].append(1)
    else:
        tip["id"].append(tip["id"][-1] + 1)
    ids.append(tip["id"][-1])
    tip["tip"].append(tip_tip)
    
new_ht = all_ht.drop(columns=["tip"])
new_ht.insert(len(new_ht.columns), "tip_id", ids, True)
new_ht.to_csv("generated/recipe_has_tip.csv", index=False, quoting=csv.QUOTE_NONNUMERIC, quotechar='"')

pd.DataFrame(tip).to_csv("generated/tip.csv", index=False, quoting=csv.QUOTE_NONNUMERIC, quotechar='"')
```

```python
# Generate recipe_has_theme
theme_data = pd.read_csv("../data/aftermath/theme.csv")
recipe_has_theme = {
    "recipe_name": [],
    "recipe_cuisine_name": [],
    "theme_name": []
}

for key in recipe_cuisine.keys():
    for rc in recipe_cuisine[key]:
        recipe_has_theme["recipe_name"].append(key)
        recipe_has_theme["recipe_cuisine_name"].append(rc)
        recipe_has_theme["theme_name"].append(random.choice(theme_data["name"]))

pd.DataFrame(recipe_has_theme).to_csv("generated/recipe_has_theme.csv", index=False, quoting=csv.QUOTE_NONNUMERIC, quotechar='"')
```

```python
# small dml generator
tables_order = ["image", "cuisine", "food_team", "ingredient", "recipe", "recipe_has_meal", "recipe_has_label",
                "tip", "recipe_has_tip", "equipment", "recipe_has_equipment", "recipe_has_ingredient",
                "theme", "recipe_has_theme", "step", "food_info", "chef", "recipe_has_chef", "chef_has_cuisine"]

queries = ["USE mydb;",
           "SET SESSION sql_mode='NO_AUTO_VALUE_ON_ZERO';"]

for table in tables_order:
    df = pd.read_csv("../data/to_import/{0}.csv".format(table))
    data = list(df.itertuples(index=False, name=None))
    
    prefix = "INSERT INTO {0} (".format(table)
    suffix = ") VALUES ("
    for col in df.columns:
        prefix = prefix+"{0},".format(col)
        if is_string_dtype(df[col]):
            suffix = suffix+"\"%s\","
        else:
            suffix = suffix+"%s,"
    sql = prefix[:-1]+suffix[:-1]+");"
    
    for t in data:
        queries.append(sql % t)    

f = open("small_dml_script.sql", "w")
for query in queries:
    f.write(query+"\n")
f.close()
```
